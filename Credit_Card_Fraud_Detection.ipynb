### Split the Dataset

```python
X = credit_card.drop('Class',axis=1)
y = credit_card[['Class']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, stratify= y, random_state=4)
```

```python
print("Training Data size --> ",X_train.shape)
print("Test Data size --> ",X_test.shape)
```

### Data Preprocessing 

```python
from sklearn.utils.class_weight import compute_sample_weight
```

```python
w_train = compute_sample_weight('balanced',y_train)
```

### Model Training

```python
from sklearn.model_selection import cross_val_score
```

### Logistic Regression 

```python
log_model = LogisticRegression()
```

```python
lW = compute_sample_weight("balanced",credit_card.Class)
```

```python
log_model.fit(X,y,sample_weight=lW)
```

```python
X_train_prediction = log_model.predict(X_train)
```

```python
log_traingScore = accuracy_score(X_train_prediction,y_train)*100
```

```python
print(f"Train Data set Accuracy {log_traingScore}")
```

```python
X_test_prediction = log_model.predict(X_test)
```

```python
log_Testscore = accuracy_score(X_test_prediction,y_test)*100
```

```python
print(f"Test Data set Accuracy {log_Testscore}")
```

```python
cross_val_score(log_model,X_train,y_train,cv = 15).mean()
```

### DecisionTree

```python
D_Tree = DecisionTreeClassifier(max_depth=3, random_state=4)
```

```python
D_Tree.fit(X_train,y_train,sample_weight=w_train)
```

```python
DTree_train_prediction = D_Tree.predict(X_train)
```

```python
dt_traingScore = accuracy_score(DTree_train_prediction,y_train)*100
```

```python
print(f"Train Data set Accuracy {dt_traingScore}")
```

```python
Dt_test_prediction = D_Tree.predict(X_test)
```

```python
dt_Testscore = accuracy_score(Dt_test_prediction,y_test)*100
```

```python
print(f"Test Data set Accuracy {dt_Testscore}")
```

### Compare both model accuracy

```python
compare = {"Logistic":{'Training SCore':log_traingScore,"TestScore" : log_Testscore},
          "DTree" : {'Training SCore':dt_traingScore,"TestScore" : dt_Testscore}
          }
```
